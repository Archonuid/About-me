### 1. Demonstrate basic date and time classes, different time formats, converting between formats, formatting dates

```python
# Import the datetime module
import datetime

# Get the current date and time
current_datetime = datetime.datetime.now()
print("Current Date and Time:", current_datetime)

# Get the current date
current_date = current_datetime.date()
print("Current Date:", current_date)

# Get the current time
current_time = current_datetime.time()
print("Current Time:", current_time)

# Create a custom date
custom_date = datetime.date(2023, 2, 24)
print("Custom Date:", custom_date)

# Create a custom time
custom_time = datetime.time(14, 30, 0)  # 2:30 PM
print("Custom Time:", custom_time)

# Combine a custom date and time into a datetime object
custom_datetime = datetime.datetime.combine(custom_date, custom_time)
print("Custom Date and Time:", custom_datetime)

# Formatting dates as strings
formatted_date = current_date.strftime("%Y-%m-%d")
print("Formatted Date:", formatted_date)

# Parsing date strings to datetime objects
date_string = "2023-02-24"
parsed_date = datetime.datetime.strptime(date_string, "%Y-%m-%d")
print("Parsed Date:", parsed_date)
```

### 2. Write a program that reads the current time and converts it to a time of day in hours, minutes and seconds plus the number of days since the epoch.

```python
# Get the current date and time
current_datetime = datetime.datetime.now()

# Extract the time components
current_time = current_datetime.time()
hours = current_time.hour
minutes = current_time.minute
seconds = current_time.second

# Calculate the number of days since the epoch
epoch = datetime.datetime(1970, 1, 1)
days_since_epoch = (current_datetime - epoch).days

# Print the results
print(f"Current Time: {hours} hours, {minutes} minutes, {seconds} seconds")
print(f"Days Since Epoch: {days_since_epoch} days")
```

### 3. Write a program to create a date from a given year, month, day and another date from a given string formats.

```python
# Create a date from year, month, and day values
year = 2023
month = 2
day = 24
date_from_values = datetime.date(year, month, day)
print("Date from values:", date_from_values)

# Create a date from a string format
date_string = "2023-02-24"
date_from_string = datetime.datetime.strptime(date_string, "%Y-%m-%d").date()
print("Date from string:", date_from_string)
```

### 4. Write a program to print a day after and before a specified date, also print the days between two given dates.

```python
# Specify the date
specified_date = datetime.date(2023, 2, 24)

# Calculate the day before and after the specified date
day_before = specified_date - datetime.timedelta(days=1)
day_after = specified_date + datetime.timedelta(days=1)

print("Specified Date:", specified_date)
print("Day Before:", day_before)
print("Day After:", day_after)

# Calculate the days between two given dates
start_date = datetime.date(2023, 2, 20)
end_date = datetime.date(2023, 2, 28)

days_between = (end_date - start_date).days

print("\nStart Date:", start_date)
print("End Date:", end_date)
print("Days Between:", days_between)
```

### 5. Write a program to create a time-series with two index labels and random values, also print the type of index.

```python
# Create a time-series with two index labels and random values
index_labels = pd.date_range(start='2023-02-24', periods=2, freq='D')
random_values = np.random.rand(2)
time_series = pd.Series(random_values, index=index_labels)

# Print the time-series
print("Time-Series:")
print(time_series)

# Print the type of index
index_type = type(time_series.index)
print("\nType of Index:", index_type)
```

### 6. Write a program to create a series of timestamps from a dataframe of integersor string columns and create a series of timestamps using specified columns.

```python
# Sample DataFrame with integer and string columns
data = {'Year': [2023, 2022, 2021],
        'Month': [1, 5, 12],
        'Day': [15, 20, 25],
        'Hour': [8, 14, 19],
        'Minute': [30, 45, 0],
        'Second': [0, 0, 0]}

df = pd.DataFrame(data)

# Create a series of timestamps using specified columns
timestamp_series = pd.to_datetime(df[['Year', 'Month', 'Day', 'Hour', 'Minute', 'Second']])

# Print the timestamp series
print("Timestamp Series:")
print(timestamp_series)
```
